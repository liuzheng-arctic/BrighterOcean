## First year ice

[0] Cold snow: Before melt onset the snow albedo is 0.85.

[1] Melting snow: Starting with the onset of snow melt, there is a linear decrease to 0.6 in 7 days.

[2] Pond formation: For the next 7 days, decrease from 0.6 to 0.32.

[3] Pond drainage: For next 7 days, increase from 0.32 to 0.54.

[4] Pond evolution: Decreases by 0.0083 d−1 until albedo reaches 0.2. Then the ice is assumed to rapidly melt and albedo drops in a single day to 0.07.

[5] Open water: Albedo of 0.07

[6] Freezeup: Once new seasonal ice begins to form from open water the albedo increases by 0.0082 d−1 to a maximum of 0.85

## Multi-year ice

[0] Before melt onset the snow albedo is 0.85.

[1] At early melt the albedo decreases to 0.81.

[2] Starting with full melt, there is a linear decrease to 0.71 in 15 days.

[3] For the next 6 days, decrease from 0.70 to 0.50.

[4] Albedo decreases by 0.0029 d–1 (but to no less than 0.2).
 - from 0.56 rather than 0.50 (Suzance's code to show pond drainage)

[5] At early freeze-up set albedo to 0.46, representing some ponds freezing.
 - set to 0.46 only if it is smaller than 0.46
